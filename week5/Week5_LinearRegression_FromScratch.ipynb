# Week 5 - Linear Regression From Scratch
## Introduction to Machine Learning
Machine Learning is a field of AI that enables systems to learn patterns from data.
Supervised Learning includes Regression and Classification.
Linear Regression is used to predict continuous values.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create Sample Dataset


```python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])
```

## Linear Regression From Scratch (Closed Form)


```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

numerator = np.sum((X - mean_x) * (Y - mean_y))
denominator = np.sum((X - mean_x) ** 2)

m = numerator / denominator
b = mean_y - m * mean_x

print('Slope:', m)
print('Intercept:', b)
```

## Gradient Descent Implementation


```python
m = 0
b = 0
learning_rate = 0.01
epochs = 100
n = len(X)

for i in range(epochs):
    Y_pred = m * X + b
    dm = (-2/n) * np.sum(X * (Y - Y_pred))
    db = (-2/n) * np.sum(Y - Y_pred)
    m = m - learning_rate * dm
    b = b - learning_rate * db

print('Gradient Descent Slope:', m)
print('Gradient Descent Intercept:', b)
```

## Visualization


```python
Y_final = m * X + b
plt.scatter(X, Y)
plt.plot(X, Y_final)
plt.title('Linear Regression From Scratch')
plt.show()
```
