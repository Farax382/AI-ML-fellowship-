# Week 7 – Unsupervised Learning & Market Basket Analysis
This notebook explores clustering, dimensionality reduction, and association rule mining using Apriori.


## Import Libraries

```python
!pip install mlxtend
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from mlxtend.frequent_patterns import apriori, association_rules
```

## Generate Dataset

```python
X, y = make_blobs(n_samples=200, centers=3, random_state=42)

plt.scatter(X[:,0], X[:,1])
plt.title('Original Dataset')
plt.show()
```

## K-Means Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.title('K-Means Clustering')
plt.show()
```

## Hierarchical Clustering

```python
model = AgglomerativeClustering(n_clusters=3)
clusters = model.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=clusters)
plt.title('Hierarchical Clustering')
plt.show()
```

## PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:,0], X_pca[:,1])
plt.title('PCA Projection')
plt.show()
```

## t-SNE Visualization

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.scatter(X_tsne[:,0], X_tsne[:,1])
plt.title('t-SNE Visualization')
plt.show()
```

## Market Basket Dataset

```python
data = {
'Milk':[1,1,0,1,0],
'Bread':[1,1,1,0,0],
'Butter':[0,1,1,1,0],
'Beer':[0,1,1,0,1]
}

df = pd.DataFrame(data)
df
```

## Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
frequent_itemsets
```

## Association Rules

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)
rules
```

## Observations
- Some items appear together frequently.
- Bread and Milk often occur in the same transaction.
- Association rule mining helps retailers understand purchasing patterns.


## Conclusion
This notebook demonstrated clustering, dimensionality reduction, and market basket analysis using Apriori.
